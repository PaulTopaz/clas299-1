## Art of devastation

Load the data set, one record for each row in the text file:

```scala
// Import the dataset. Create a Vector of strings after converting each record into its own line.


val url = "https://raw.githubusercontent.com/neelsmith/nomisma/master/cex/devastation-2010-04-10.cex"

import scala.io.Source
val lines = Source.fromURL(url).getLines.toVector
```

```scala
// Split each data string into columns using the split method. 


val withColumns = lines.map( ln => ln.split("#").toVector)
```

```scala
// Define a class called AoDObverse and include the following columns: aod number,  authority, obvType, obvLegend, and startdate. Define an expression named Pretty and set it to create a string with those 5 properties.
// This will give you the only information that you want when you are searching a term later on.


case class AoDObverse ( aod: String,  
                       authority: String, 
                       obvType: String,  
                       obvLegend: String,  
                       startDate: String) {
  
  def pretty: String = {
    aod + "\nAuthority: " + authority + "\nType: " + obvType + "\nLegend: " + obvLegend + "\nDate: " + startDate
  }
}
```

```scala
// Using the mapping method,  extract the 7th column from each vector to give us the text on the Obverse side of the medallions. 


val obvDescriptions = withColumns.map(v => v(7))
```

```scala
// Define an expression named aod and use the filter method to filter out the columns that we do not want to use.


val aod = withColumns.filter(v => v.size >= 14).map(v => AoDObverse(v(1), v(4), v(7), v(8), v(13)))

```

```scala
// Use the println method to see how this will look.


println(aod.tail.head.pretty)
```

```scala
// To avoid uppercase letters and punctuation affecting our results, change all the letters to lowercase and replace all the characters that are not included in “^a-z” with a space. Create one string of all the obverse side text when you have done so.  


val allObvText = obvDescriptions.mkString(" ").toLowerCase.replaceAll("[^a-z ]","")
```

```scala
// Create a list of stop words to filter out from the string above. 


val stopWords = Vector("to", "of", "and", "a", "l", "with", "in", "r", "on", "the")
```

```scala
// Create a Vector of strings after filtering out the stop words and split each word into its own column.


val filteredWords = allObvText.split(" ").filterNot(wd => stopWords.contains(wd)).toVector
```

```scala
// Group the words that are the same using the groupBy method and count the number of words per group using the map method and v.size.


val wordsGrouped = filteredWords.groupBy( w => w).toVector
val wordCount = wordsGrouped.map{ case (wd, v) => (wd, v.size)}
```

```scala
// Sort these words in alphabetical order using the sortBy and reverse method.


val sortedCounts = wordCount.sortBy(_._2).reverse
```

```scala
// Print these results to then analyze the most frequently used words and the least frequently used words. 
// These words will be the top 10 and the bottom 10 of the list.

println(sortedCounts.mkString("\n"))
```
